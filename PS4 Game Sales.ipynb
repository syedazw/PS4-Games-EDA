# Cleaning

### Remove null values from *Year & Publisher* column

```python
# remove null values using interpolate technique - year column

ps4_data['Year'] = ps4_data['Year'].interpolate(method='linear').round()
ps4_data
```

##### - numerical data kai lie -mean, median
##### - categorical data kai lie -mode

```python
# remove null values by mode - publisher column 

pd.set_option('display.max_rows', None)

ps4_data['Publisher'].fillna(ps4_data['Publisher'].mode()[0], inplace=True)
ps4_data
```

### Clean the values of Regions columns

```python
# imputation for 0 values rows

# group wise imputation for rows where Regions are 0
columns_to_check = ['North America', 'Europe', 'Japan', 'Rest of World', 'Global']
rows_to_impute = (ps4_data[columns_to_check]==0.0).all(axis=1)

# apply imputation
for col in columns_to_check:
    columns_median = ps4_data[col].replace(0.0, np.nan).median()

    # generate random values near the mean
    ran_val = np.random.uniform(
        max(0,columns_median-0.1), min(2.1, columns_median+0.1), size= rows_to_impute.sum())
    ps4_data.loc[rows_to_impute, col] = ran_val

ps4_data

#debugging for correctness in data where all rows have 0 values in specific 5 columns

# print("Rows to Impute (True/False):")
# print(rows_to_impute)
# print("Total Rows to Impute:", rows_to_impute.sum())

# print("Data in Columns Before Imputation:")
# print(ps4_data[columns_to_check])

# print("Generated Random Values for Imputation:")
# print(ran_val)
```

#### 8. Compare the sales growth trend for "Action" and "Shooter" genres over the years.

```python
# filter genre column
fil_gen = ps4_data[ps4_data['Genre'].isin(['Action', 'Shooter'])]

gen_year = fil_gen.groupby(['Year','Genre'])['Global'].sum().unstack()

gen_year.plot(kind='line', figsize=(8,5))
plt.title('Sales Growth Trend for "Action" and "Shooter" Genres Over the Years')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.grid()
plt.legend(title='Genre')
plt.show()
```
